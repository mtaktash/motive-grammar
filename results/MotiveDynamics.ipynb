```python
import os
import pandas as pd


OUTPUT_FOLDER = '/Users/maria/github/motive-grammar/results'
MOTIVES = [
    'laugh',
    'transport',
]

COLORS = [
    'blue',
    'green',
]

BOOKS = [
    'a_hero_of_our_time',
    'anna_karenina',
    'crime_and_punisnment',
    'master_and_margarita',
    'war_and_peace',
]
```

# Разные мотивы в книге

```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

def plot_motives(bookname):
    for motive, color in zip(MOTIVES, COLORS):
        file_path = f'{OUTPUT_FOLDER}/{motive}/{bookname}.tsv'
        data = pd.read_csv(file_path, sep='\t')
        num_bins = max(data.text) // 10
        
        plt.figure(figsize=(12,2))
        plt.title(f'{motive} in {bookname}')
        sns.distplot(data.text, bins=num_bins, hist=True, color=color);
```

```python
plot_motives('master_and_margarita')
```

# Мотив в разных книгах

```python
def plot_motive_dynamics(motive):
    for bookname in BOOKS:
        file_path = f'{OUTPUT_FOLDER}/{motive}/{bookname}.tsv'
        data = pd.read_csv(file_path, sep='\t')
        num_bins = max(data.text) // 10
        
        plt.figure(figsize=(12,2))
        plt.title(f'{motive} in {bookname}')
        sns.distplot(data.text, bins=num_bins, hist=True);
```

```python
plot_motive_dynamics('laugh')
```
