```python
import os
import pandas as pd


OUTPUT_FOLDER = '/Users/maria/github/motive-grammar/laugh/output'

all_data = {}
for filename in os.listdir(OUTPUT_FOLDER):
    if filename.endswith('.tsv'):
        print(filename)
        file_path = f'{OUTPUT_FOLDER}/{filename}'
        data = pd.read_csv(file_path, sep='\t')
        all_data[filename[:-4]] = data
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
for key in all_data:
    data = all_data[key]
    num_bins = max(data.text) // 5
    plt.figure(figsize=(10,5))
    plt.title(key)
    sns.distplot(data.text, bins=num_bins, hist=True);
```
