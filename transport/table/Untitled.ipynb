```python
from bs4 import BeautifulSoup
import glob
from tqdm import tqdm
import string
import re
```

```python
with open('out.html', 'r') as output:
    file_str = ''.join(output.readlines())
    soup = BeautifulSoup(file_str, 'lxml')
```

```python
def prepare_string(a):
    t = a.replace('\t', '').replace('\n', '')\
        .replace('\xa0', '').replace(' ,', '').replace(' .', '').replace(' ?', '')\
        .replace(' !', '').replace(' :', '').replace(' (', '').replace(' )', '')\
        .lower().translate(str.maketrans('', '', '–—.,?!()…«»:;')).strip()
    return re.sub(' +', ' ', t)
```

```python
motifs = [prepare_string(soup.select_one('a[name="{}"]'.format(tag['href'][1:])).text)[:-4] 
          for tag in tqdm(soup.select('td > a'))]
```

```python
motifs
```

```python
for filename in glob.glob('2206/*.txt'):
    q = 0
    with open(filename) as file:
        text = prepare_string(file.read())
        for motif in motifs:
            if motif in text:
                q += 1
        print('{}\t{}'.format(filename, q))
        
```
